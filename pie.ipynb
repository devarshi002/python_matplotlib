# pie plot 

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["c", "c++", "java", "python"]
ex = [0.2,0.0,0.0,0.0]

plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=True, startangle=0, )
plt.title("languages")
plt.legend()
plt.show()
```

```python
p = [10,20,30,40]
q = [40,30,20,10]
u = ["c", "c++", "java", "python"]

plt.pie(p, labels=u, radius=1.5)
plt.pie([1],colors="w")
plt.show()
```
