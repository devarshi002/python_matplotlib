# python bar plot

```python
import matplotlib.pyplot as plt
x = ["python", "c", "java", "c++"]
y = [85,60,75,40]
plt.xlabel("languages", fontsize=15)
plt.ylabel("no.", fontsize=15)
plt.title("analysis", fontsize=15)
plt.bar(x, y, width=0.4, color="r", label="popularity")
plt.legend()
plt.show()
```
