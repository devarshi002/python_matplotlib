```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1

plt.hist([data1, data2], bins=30, color=['blue', 'green'], alpha=0.7, label=['Data 1', 'Data 2'], edgecolor='black', linewidth=1.2, histtype='barstacked')

# Add title and labels
plt.title("Comparison of Two Data Sets", fontsize=15)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Add a legend
plt.legend()

# Show grid
plt.grid(True)

# Show plot
plt.show()
```
